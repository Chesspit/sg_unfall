```python
df = df.iloc[:, [0, 2, 7, 11, 12, 13, 15, 21, 25, 26, 33, 37]]
df.head()
```

```python
column_name_mapping = {
    'AccidentUID': 'ID',
    'AccidentType_de': 'Unfalltyp',
    'AccidentSeverityCategory_de': 'Unfallschwere',
    'AccidentInvolvingPedestrian': 'Fussgänger',
    'AccidentInvolvingBicycle': 'Fahrrad',
    'AccidentInvolvingMotorcycle': 'Motorrad',
    'RoadType_de': 'Strassentyp',
    'Location': 'Location',
    'AccidentYear': 'Jahr',
    'AccidentMonth': 'Monat',
    'AccidentWeekDay_de': 'Wochentag',
    'AccidentHour': 'Stunde'
}
```

```python
df.rename(columns=column_name_mapping, inplace=True)
```

```python
df[['Breitengrad', 'Längengrad']] = df['Location'].str.split(',', expand=True).astype(float)
```

```python
df.drop(columns=['Location'], inplace=True)
```

```python
df['Datum'] = pd.to_datetime(df['Jahr'].astype(str) + '-' + df['Monat'].astype(str) + '-01')
```
